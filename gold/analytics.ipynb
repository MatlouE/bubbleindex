```python
fundamentals = pd.read_csv('/workspaces/bubbleindex/silver/combined_fundamentals.csv')
prices = pd.read_csv('/workspaces/bubbleindex/silver/combined_stockprices.csv')
```

```python
fundamentals['Date'] = pd.to_datetime(fundamentals['Date'])
prices['Date'] = pd.to_datetime(prices['Date'], utc='True')
```

Fundamentals Analysis

```python
#get latest fundamentals & prices for each company
latest_fundamentals = fundamentals.sort_values('Date').groupby('ticker').tail(1).reset_index(drop=True)
latest_prices = prices.sort_values('Date').groupby('ticker').tail(1).reset_index(drop=True)
```

```python
fund_trends = fundamentals.sort_values(['ticker', 'Date']).copy()
```

```python
fund_trends
```

```python
print(sorted(fund_trends['ticker'].unique()))
```

Profitability trend of each company (Net Margin %)

percentage change for each company 

eg. google[ending net margin] -  [start net margin]
for a metric measurement or like comparions u would prefer to
see the relative change then just numbers
        

```python
print('Net Margin:\n')
for ticker in sorted(fund_trends['ticker'].unique()):
    ticker_data = fund_trends[fund_trends['ticker'] == ticker]
    ticker_data = ticker_data[['Date', 'Net Margin']].copy()
    ticker_data['Net Margin'] = ticker_data['Net Margin'] * 100
    ticker_data = ticker_data.sort_values('Date')
    
    #if ticker_data has more than 1 entry then calculate the change in net margin from the first to the last entry 
    if len(ticker_data) > 1:
        margin_change = ticker_data['Net Margin'].iloc[-1]  - ticker_data['Net Margin'].iloc[0]
        print(f"{ticker} Net Margin Change: {ticker_data['Net Margin'].iloc[-1]:.1f}% Current Margin | {margin_change:.1f}% change")
    else:
        print(f'{ticker} Net Margin: {ticker_data["Net Margin"].values[0]:.1f}% (limited data)')

print("\nRevenue Growth Trends:\n")
for ticker in sorted(fund_trends['ticker'].unique()):
    ticker_data = fund_trends[fund_trends['ticker'] == ticker]
    ticker_data = ticker_data[['Date', 'Revenue growth']].copy()
    ticker_data = ticker_data.sort_values('Date')
    rev_growth_values = ticker_data['Revenue growth'].dropna() * 100

    if len(rev_growth_values) > 0:
        avg_growth = rev_growth_values.mean()
        latest_growth = rev_growth_values.iloc[-1 ] if len(rev_growth_values) > 0 else 0

        print(f"{ticker} average revenue growth : {avg_growth:.1f}%| Current Revenue Growth: {latest_growth:.1f}%")
    else:
        print(f"{ticker} Revenue Growth: No data available")

print('\nR&D trends:\n')
for ticker in sorted(fund_trends['ticker'].unique()):
    ticker_data = fund_trends[fund_trends['ticker'] == ticker].copy()
    ticker_data = ticker_data.sort_values('Date')
    rnd_growth = ticker_data[['Date', 'R&D to Revenue']]
    rnd_growth = ticker_data['R&D to Revenue'] * 100

    if len(rnd_growth) > 0:
        avg_rnd_growth = rnd_growth.mean()
        latest_rnd_growth = rnd_growth.iloc[-1] if len(rnd_growth) > 0 else 0

        print(f"{ticker} Average R&D growth: {avg_rnd_growth:.1f}% | Latest R&D growth: {latest_rnd_growth:.1f}%")
    else: 
        print(f"{ticker} R&D Growth: No data available")
```
