Market Data

```python
##historical market data
nvda_price = nvda.history(period='6y')
nvda_price = nvda_price[['Close', 'Volume']].reset_index()
nvda_price.rename(columns={'Close' : 'close_price'}, inplace=True)
print(nvda_price.head())

#save df to excel and csv --- IGNORE ---
nvda_price.to_csv('nvda_price.csv', index=False)
```

Balance Sheet

```python
#Balance Sheet

nvda_balancesheet = nvda.balance_sheet.T
print(type(nvda_balancesheet))
print(nvda_balancesheet.head())

#save df to excel and csv --- IGNORE ---
nvda_balancesheet.to_csv('nvda_balancesheet.csv', index=True)
```
