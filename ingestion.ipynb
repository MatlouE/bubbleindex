```python
import yfinance as yf
import pandas as pd
from matplotlib import pyplot as plt
```

```python

nvda = yf.Ticker("NVDA")
```

```python
##historical market data
nvda_price = nvda.history(period='6y')
nvda_price = nvda_price[['Close', 'Volume']].reset_index()
nvda_price.rename(columns={'Close' : 'close_price'}, inplace=True)
print(nvda_price.head())
```

```python
#Fundamentals

fundamentals = nvda.get_balancesheet()
print(type(fundamentals))
print(fundamentals.head())
```

```python
#save df to excel

fundamentals.to_excel('nvda_fundamentals.xlsx', index=False)
```
